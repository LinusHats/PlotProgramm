```python
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use("https://raw.githubusercontent.com/LinusHats/PlotProgramm/main/ownstyle.mplstyle")
```

## Festlegen der ganzen Texte...und der Daten
hier werden erst die Daten eingelesen. Am einfachsten funktioniert das mit Pandas, also den Pfad zur Exeltabelle angeben und bei x_/y_werte den Namen der Spalte eigeben, in dem die Daten zu finden sind. <br>
sonst noch wichtig: <br>
Bei x_label zu sehen es kann LaTeX in den Schriften verwendet werden. Dazu: ein "r" vor die Anführungszeichen und dann wie immer die Befehle in "$$" <br>
Bei den Farben ist einiges möglich...am besten einfach Klarnamen("green""red""blue") oder hexadecimal ("#ff0000" (rot), "#00ff00" (gelb), "#0000ff" (blau)) eingeben

```python
dateipfad = "PIC309"                                           #Hier den Dateipfad eigeben
x_werte = "Wellenlänge"                                        #Hier den Namen der Spalte mit den x_werten eingeben
y_werte = "Intensitaet"                                        #Hier den Namen der Spalte mit den y_werten eingeben
Überschrift = "Anregungsspektrum Picolinsäure"                 #Gib dem Diagramm einen Namen (Üebrschrift)
y_label = "Anregung [counts]"                                  #Gib den y-Werten einen Namen (y_label)
x_label = r"Wellenlänge ($\lambda$) [nm]"                      #Gib den x-Werten einen namen (y_label)
legend = "Messwerte"                                           #Hier den legendeneintrag festlegen 
color = "Tab:orange"                                           #Möglichkeiten sind ganze worte od. hexdacimal od. einzelne Buchstaben zB r für Rot oder g für Grün
```

```python
#-----------------------Hier werden die Daten eingelesen--------------------------------------#
Daten = pd.read_excel('{}.xlsx'.format(dateipfad))
x = Daten["{}".format(x_werte)]
y = Daten["{}".format(y_werte)]
#----------------------Zur Überprüfung--------------------------------------------------------#
Daten
```

## das eigentliche Plotten
Hier wird dann noch geplottet...Man kann alles von oben auch direkt hier eintragen, so finde ich's aber übersichtlicher<br>
Für plots mit mehr linien kann man den ganzen bums analog machen, einfach plt.plot(x, andere y-werte,label="mein label",color="neue Farbe")

```python
#---------------------und jetzt noch die Daten plotten----------------------------------------#
plt.plot(x,y, label=legend, color=color)
plt.title(Überschrift)
plt.ylabel(y_label)
plt.xlabel(x_label)
#um die limits der Achsen zu begrenzen: plt.xlim((min,max)) bzw plt.ylim((min,max))
plt.legend(loc=2, bbox_to_anchor=(1,1))      #Falls man einen Titel über der legende Will: ,label="Mein Label"
#plt.xlim((min,max))                         #Falls begrenzungen der Achsen gewünscht einfach die passenden Zahlen eintragen
#plt.ylim((min,max))
plt.show()
```
